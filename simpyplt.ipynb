```python
import pandas as pd
import matplotlib.pyplot as plt

# 用你目前的三組平均值建立資料表
bar_data = pd.DataFrame({
    "scenario": [
        "HR:1, Interviewer:1",
        "HR:2, Interviewer:1",
        "HR:1, Interviewer:2"
    ],
    "mean_process_time": [
        76.07,
        71.51,
        62.30
    ]
})

# 設定中文字型（Windows：微軟正黑體，macOS 可用 PingFang TC）
plt.rcParams['font.family'] = 'Microsoft JhengHei'
plt.rcParams['axes.unicode_minus'] = False

# 畫出平均處理時間比較圖（繁體版）
bar_data.set_index("scenario").plot.bar(legend=False, color="skyblue")
plt.ylabel("平均處理時間（分鐘）")
plt.title("不同人力配置下的履歷處理時間比較")
plt.xticks(rotation=15)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
